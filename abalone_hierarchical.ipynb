```python
#label_encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
df['Sex']=a.fit_transform(df['Sex'])
df
```

```python
#heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,15))
corr=df.corr()
sns.heatmap(corr,annot=True,fmt='.2g',linewidth=1)
plt.tight_layout()
plt.show()
```

```python
#dropping_table
df=df.drop('Sex',axis=1)
df
```

```python
df1=df.values
```

```python
#Dendogram
import scipy.cluster.hierarchy as h
plt.figure(figsize=(20,15))
h.dendrogram(h.linkage(df,method='ward'))
plt.show()
```

```python
#AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
y_hc=hc.fit_predict(df1)
y_hc
```

```python
#visualizing_the_clusters_using_scatterplot
plt.figure(figsize=(15,7))
plt.scatter(df1[y_hc==0,5],df1[y_hc==0,6],color='green',label='cluster1',s=50)
plt.scatter(df1[y_hc==1,5],df1[y_hc==1,6],color='orange',label='cluster2',s=50)
plt.scatter(df1[y_hc==2,5],df1[y_hc==2,6],color='blue',label='cluster3',s=50)
plt.scatter(df1[y_hc==3,5],df1[y_hc==3,6],color='yellow',label='cluster4',s=50)
plt.scatter(df1[y_hc==4,5],df1[y_hc==4,6],color='grey',label='cluster5',s=50)
plt.grid(False)
plt.title('Number of clusters')
plt.xlabel('viscera weight')
plt.ylabel('shell weight')
plt.show()
```

```python
#adding_column_clusters
df['Clusters']=pd.DataFrame(y_hc)
df
```
